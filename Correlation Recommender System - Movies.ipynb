```python
rating_count = pd.DataFrame(rating_frame.groupby('movieId')['rating'].count())
```

```python
movies_frame[movies_frame['movieId'] == 356]
```

```python
movies_crosstab = pd.pivot_table(data=rating_frame, index='userId', columns='movieId', values='rating')
```

```python
forrestGump = movies_crosstab[356]
```

```python
forrestGump[forrestGump>0]
```

```python
similar = movies_crosstab.corrwith(forrestGump)
similar
```

```python
PearsonR = pd.DataFrame(similar, columns=['PerasonR'])
```

```python
PearsonR.dropna(inplace=True)
```

```python
PearsonR
```

```python
summary = PearsonR.join(rating_count['rating'])
summary.head(10)
```

```python
summaryCorr = summary[summary['rating']>=10].sort_values('PerasonR', ascending = False)
```
